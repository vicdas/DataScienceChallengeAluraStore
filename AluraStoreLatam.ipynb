### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
# Para calcular el total de una tienda Uno
# Primero calcular la facturación por fila
tienda['Facturación'] = tienda['Precio'] + tienda['Costo de envío']

# Posteriormrnte calcular el total de facturación
facturacion_total_T1 = tienda['Facturación'].sum()

#Desplegar el total de facturación de la tienda
print(f"Facturación total de la tienda 1: {facturacion_total_T1}")

# Para calcular el total de una tienda Dos
# Primero calcular la facturación por fila
tienda2['Facturación'] = tienda2['Precio'] + tienda2['Costo de envío']

# Posteriormrnte calcular el total de facturación
facturacion_total_T2 = tienda2['Facturación'].sum()

#Desplegar el total de facturación de la tienda
print(f"Facturación total de la tienda 2: {facturacion_total_T2}")

# Para calcular el total de una tienda Tres
# Primero calcular la facturación por fila
tienda3['Facturación'] = tienda3['Precio'] + tienda3['Costo de envío']

# Posteriormrnte calcular el total de facturación
facturacion_total_T3 = tienda3['Facturación'].sum()

#Desplegar el total de facturación de la tienda
print(f"Facturación total de la tienda 3: {facturacion_total_T3}")

# Para calcular el total de una tienda Cuatro
# Primero calcular la facturación por fila
tienda4['Facturación'] = tienda4['Precio'] + tienda4['Costo de envío']

# Posteriormrnte calcular el total de facturación
facturacion_total_T4 = tienda4['Facturación'].sum()

#Desplegar el total de facturación de la tienda
print(f"Facturación total de la tienda 4: {facturacion_total_T4}")
```

# 2. Ventas por categoría

```python
# Agrupar por tienda y categoría, y contar cuántas ventas hubo
ventas_por_tienda_categoria = tienda.groupby(['Producto','Categoría del Producto']).size().reset_index(name='Cantidad')

# Ordenar para ver las categorías más populares por tienda
ventas_ordenadas = ventas_por_tienda_categoria.sort_values(['Cantidad','Categoría del Producto'], ascending=[False, True]).reset_index(drop=True).head(5)

#Desplegar las ventas por producto y categoría del producto
print("Cantidad de productos de las categorías más populares vendidos en la tienda 1:")
print(ventas_ordenadas.to_string(index=False))

# Agrupar por tienda y categoría, y contar cuántas ventas hubo
ventas_por_tienda_categoria = tienda2.groupby(['Producto','Categoría del Producto']).size().reset_index(name='Cantidad')

# Ordenar para ver las categorías más populares por tienda
ventas_ordenadas = ventas_por_tienda_categoria.sort_values(['Cantidad','Categoría del Producto'], ascending=[False, True]).reset_index(drop=True).head(5)

#Desplegar las ventas por producto y categoría del producto
print("Cantidad de productos de las categorías más populares vendidos en la tienda 2:")
print(ventas_ordenadas.to_string(index=False))

# Agrupar por tienda y categoría, y contar cuántas ventas hubo
ventas_por_tienda_categoria = tienda3.groupby(['Producto','Categoría del Producto']).size().reset_index(name='Cantidad')

# Ordenar para ver las categorías más populares por tienda
ventas_ordenadas = ventas_por_tienda_categoria.sort_values(['Cantidad','Categoría del Producto'], ascending=[False, True]).reset_index(drop=True).head(5)

#Desplegar las ventas por producto y categoría del producto
print("Cantidad de productos de las categorías más populares vendidos en la tienda 3:")
print(ventas_ordenadas.to_string(index=False))

# Agrupar por tienda y categoría, y contar cuántas ventas hubo
ventas_por_tienda_categoria = tienda4.groupby(['Producto','Categoría del Producto']).size().reset_index(name='Cantidad')

# Ordenar para ver las categorías más populares por tienda
ventas_ordenadas = ventas_por_tienda_categoria.sort_values(['Cantidad','Categoría del Producto'], ascending=[False, True]).reset_index(drop=True).head(5)

#Desplegar las ventas por producto y categoría del producto
print("Cantidad de productos de las categorías más populares vendidos en la tienda 4:")
print(ventas_ordenadas.to_string(index=False))
```

# Gráficos de cada sección

```python
import matplotlib.pyplot as plt

# Datos de facturación
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
facturacion = [1212258300.0, 1175828600.0, 1156536200.0, 1093693100.0]

# Determinar índice de la tienda con mayor facturación
indice_max = facturacion.index(max(facturacion))

# Colores: verde para la mayor, azul para el resto
colores = ['#4CAF50' if i == indice_max else '#2196F3' for i in range(len(tiendas))]

# Crear gráfico
plt.figure(figsize=(10, 6))
barras = plt.bar(tiendas, facturacion, color=colores)

# Título y etiquetas
plt.title('Facturación Total por Tienda', fontsize=16)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Facturación (en pesos)', fontsize=12)

# Mostrar valores y destacar la mayor facturación
for i, barra in enumerate(barras):
    yval = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, yval + 1e7, f'{yval:,.0f}', ha='center', fontsize=10)

    # Etiqueta especial para la tienda con mayor facturación
    if i == indice_max:
        plt.text(barra.get_x() + barra.get_width()/2, yval + 5e7, 'Mayor facturación',
                 ha='center', fontsize=12, color='green', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos de calificaciones promedio por tienda
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
calificaciones = [3.98, 4.04, 4.05, 4.00]

# Determinar índice de la tienda con mayor calificación
indice_max = calificaciones.index(max(calificaciones))

# Colores: verde para la mayor, azul para el resto
colores = ['#4CAF50' if i == indice_max else '#2196F3' for i in range(len(tiendas))]

# Crear gráfico
plt.figure(figsize=(10, 6))
barras = plt.bar(tiendas, calificaciones, color=colores)

# Título y etiquetas
plt.title('Calificación Promedio de los Clientes por Tienda', fontsize=16)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Calificación Promedio', fontsize=12)
plt.ylim(3.9, 4.1)  # Ajuste visual para que se vea mejor

# Mostrar valores y destacar la mayor calificación
for i, barra in enumerate(barras):
    yval = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, yval + 0.005, f'{yval:.2f}', ha='center', fontsize=10)

    if i == indice_max:
        plt.text(barra.get_x() + barra.get_width()/2, yval + 0.02, 'Mayor calificación',
                 ha='center', fontsize=12, color='green', fontweight='bold')

plt.tight_layout()

# Guardar gráfico en PNG y JPG
plt.savefig("calificacion_promedio_tiendas.png")
plt.savefig("calificacion_promedio_tiendas.jpg")

# Mostrar gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos de costo promedio de envío por tienda
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
envios_promedio = [26018.61, 25216.24, 24805.68, 23459.46]

# Determinar la tienda con mayor costo promedio
indice_max = envios_promedio.index(max(envios_promedio))

# Colores: rojo para la mayor, azul para las demás
colores = ['#E53935' if i == indice_max else '#42A5F5' for i in range(len(tiendas))]

# Crear la gráfica
plt.figure(figsize=(10, 6))
barras = plt.bar(tiendas, envios_promedio, color=colores)

# Añadir títulos y etiquetas
plt.title('Costo Promedio de Envío por Pedido por Tienda', fontsize=16)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Costo Promedio de Envío (en pesos)', fontsize=12)

# Mostrar los valores sobre las barras
for i, barra in enumerate(barras):
    yval = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, yval + 200, f'{yval:,.2f}', ha='center', fontsize=10)

    if i == indice_max:
        plt.text(barra.get_x() + barra.get_width()/2, yval + 1200, 'Mayor costo promedio',
                 ha='center', fontsize=12, color='red', fontweight='bold')

plt.tight_layout()

# Guardar gráfico como PNG y JPG
plt.savefig("envio_promedio_tiendas.png")
plt.savefig("envio_promedio_tiendas.jpg")

# Mostrar gráfico
plt.show()
```
