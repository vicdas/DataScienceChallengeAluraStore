# Gráfico de Facturación total

```python
import matplotlib.pyplot as plt

# Datos de facturación
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
facturacion = [1212258300.0, 1175828600.0, 1156536200.0, 1093693100.0]

# Determinar índice de la tienda con mayor facturación
indice_max = facturacion.index(max(facturacion))

# Colores: verde para la mayor, azul para el resto
colores = ['#4CAF50' if i == indice_max else '#2196F3' for i in range(len(tiendas))]

# Crear gráfico
plt.figure(figsize=(10, 6))
barras = plt.bar(tiendas, facturacion, color=colores)

# Título y etiquetas
plt.title('Facturación Total por Tienda', fontsize=16)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Facturación (en pesos)', fontsize=12)

# Mostrar valores y destacar la mayor facturación
for i, barra in enumerate(barras):
    yval = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, yval + 1e7, f'{yval:,.0f}', ha='center', fontsize=10)

    # Etiqueta especial para la tienda con mayor facturación
    if i == indice_max:
        plt.text(barra.get_x() + barra.get_width()/2, yval + 5e7, 'Mayor facturación',
                 ha='center', fontsize=12, color='green', fontweight='bold')

plt.tight_layout()
plt.show()
```

# Gráfico de calificaciones promedio

```python
import matplotlib.pyplot as plt

# Datos de calificaciones promedio por tienda
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
calificaciones = [3.98, 4.04, 4.05, 4.00]

# Determinar índice de la tienda con mayor calificación
indice_max = calificaciones.index(max(calificaciones))

# Colores: verde para la mayor, azul para el resto
colores = ['#4CAF50' if i == indice_max else '#2196F3' for i in range(len(tiendas))]

# Crear gráfico
plt.figure(figsize=(10, 6))
barras = plt.bar(tiendas, calificaciones, color=colores)

# Título y etiquetas
plt.title('Calificación Promedio de los Clientes por Tienda', fontsize=16)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Calificación Promedio', fontsize=12)
plt.ylim(3.9, 4.1)  # Ajuste visual para que se vea mejor

# Mostrar valores y destacar la mayor calificación
for i, barra in enumerate(barras):
    yval = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, yval + 0.005, f'{yval:.2f}', ha='center', fontsize=10)

    if i == indice_max:
        plt.text(barra.get_x() + barra.get_width()/2, yval + 0.02, 'Mayor calificación',
                 ha='center', fontsize=12, color='green', fontweight='bold')

plt.tight_layout()

# Guardar gráfico en PNG y JPG
plt.savefig("calificacion_promedio_tiendas.png")
plt.savefig("calificacion_promedio_tiendas.jpg")

# Mostrar gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Datos organizados
data = {
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Producto más vendido": [60, 65, 57, 62],
    "Producto menos vendido": [33, 32, 35, 33]
}

productos = {
    "Tienda 1": ["Microondas", "Auriculares con micrófono"],
    "Tienda 2": ["Iniciando en programación", "Juego de mesa"],
    "Tienda 3": ["Kit de bancas", "Bloques de construcción"],
    "Tienda 4": ["Cama box", "Guitarra eléctrica"]
}

df = pd.DataFrame(data)

# Gráfico de barras agrupadas
x = range(len(df))
bar_width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar([i - bar_width/2 for i in x], df["Producto más vendido"], width=bar_width, label="Producto más vendido", color="mediumseagreen")
bars2 = ax.bar([i + bar_width/2 for i in x], df["Producto menos vendido"], width=bar_width, label="Producto menos vendido", color="salmon")

# Etiquetas y leyenda
ax.set_xticks(x)
ax.set_xticklabels(df["Tienda"])
ax.set_ylabel("Cantidad de ventas")
ax.set_title("Productos más y menos vendidos por tienda")
ax.legend()

# Agregar nombres de productos encima de las barras
for i, tienda in enumerate(df["Tienda"]):
    ax.text(i - bar_width/2, df["Producto más vendido"][i] + 1, productos[tienda][0], ha='center', fontsize=8, rotation=90)
    ax.text(i + bar_width/2, df["Producto menos vendido"][i] + 1, productos[tienda][1], ha='center', fontsize=8, rotation=90)

plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos de costo promedio de envío por tienda
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
envios_promedio = [26018.61, 25216.24, 24805.68, 23459.46]

# Determinar la tienda con mayor costo promedio
indice_max = envios_promedio.index(max(envios_promedio))

# Colores: rojo para la mayor, azul para las demás
colores = ['#E53935' if i == indice_max else '#42A5F5' for i in range(len(tiendas))]

# Crear la gráfica
plt.figure(figsize=(10, 6))
barras = plt.bar(tiendas, envios_promedio, color=colores)

# Añadir títulos y etiquetas
plt.title('Costo Promedio de Envío por Pedido por Tienda', fontsize=16)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Costo Promedio de Envío (en pesos)', fontsize=12)

# Mostrar los valores sobre las barras
for i, barra in enumerate(barras):
    yval = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, yval + 200, f'{yval:,.2f}', ha='center', fontsize=10)

    if i == indice_max:
        plt.text(barra.get_x() + barra.get_width()/2, yval + 1200, 'Mayor costo promedio',
                 ha='center', fontsize=12, color='red', fontweight='bold')

plt.tight_layout()

# Guardar gráfico como PNG y JPG
plt.savefig("envio_promedio_tiendas.png")
plt.savefig("envio_promedio_tiendas.jpg")

# Mostrar gráfico
plt.show()
```

# Conclusión

Recomendando al señor Juan, la Tienda 3 emerge como la opción más sólida para la venta, a pesar de no tener la facturación más alta.

Justificación:

Máxima Satisfacción del Cliente: La Tienda 3 ostenta la calificación promedio de clientes más alta (4.05). Un alto nivel de satisfacción del cliente es un activo invaluable para cualquier negocio, ya que se traduce en lealtad, recomendaciones y una base de clientes sólida. Esto puede hacer que la tienda sea más atractiva para un comprador a largo plazo.

Menor Costo de Envío: El costo promedio de envío por pedido en la Tienda 3 es el más bajo de todas las tiendas (24805.68). Los bajos costos operativos son un factor clave para la rentabilidad y hacen que la tienda sea más eficiente y atractiva para un potencial comprador.

Enfoque en Muebles: La Tienda 3 muestra una concentración en la categoría de muebles dentro de sus productos más vendidos. Si el mercado de muebles es estable o en crecimiento, esto podría ser un punto fuerte para la tienda.

Si bien la Tienda 1 tiene la facturación total más alta, su baja calificación de clientes y su alto costo de envío son señales de posibles problemas subyacentes que podrían preocupar a un comprador. Vender la Tienda 3, con su clientela satisfecha y costos de envío optimizados, podría resultar en una venta más fluida y a un precio competitivo, destacando la salud y la eficiencia operativa del negocio.

Consideración adicional: Si el señor Juan prioriza absolutamente la mayor facturación inmediata y está dispuesto a abordar las preocupaciones sobre la satisfacción del cliente y los costos de envío, la Tienda 1 podría ser una opción. Sin embargo, esto requeriría una comunicación transparente con el comprador sobre estos aspectos.

En última instancia, considerando un equilibrio entre el atractivo para el comprador y la salud operativa del negocio, la Tienda 3 se presenta como la recomendación más prudente para el señor Juan.
